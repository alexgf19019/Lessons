```python
import numpy as np
import scipy
import scipy.linalg
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def funWeierstrass(x):
    #cos(3*x*pi)/2 + cos(3**2*x*pi)/2**2 + ... + cos(3**n*x*pi)/2**n
    return (np.cos(3*x*np.pi) / 2 + np.cos(3**2*x*np.pi) / 2**2 + np.cos(3*x**3*np.pi) / 2**3)
    #return np.cos(3*x*math.pi)
```

```python
MAX_X = 10
x = np.arange(MAX_X)
y = funWeierstrass(x)
plt.plot(x, y)
plt.xlabel('x')   # string must be enclosed with quotes '  '
plt.ylabel('y')
plt.title('Weierstrass Function')
plt.legend(['Weierstrass Function']) 
plt.show()
```

```python
def HilbertSpace(x):
#    return 1/x
    return 1/(np.sqrt(x))
```

```python

def FibNumber(N):
    """
        Fibonachi numbers definition function.
    """
    return (1/np.sqrt(5))*(((1+np.sqrt(5))/2)**N - ((1-np.sqrt(5))/2)**N)
```

```python
print (FibNumber(10))
```
